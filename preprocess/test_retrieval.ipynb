```python
# Pinecone Retrieval 테스트
from pinecone_retrieval import embed_text, retrieve_from_pinecone
```

```python
# 1. 텍스트 임베딩 테스트
text = "백엔드 개발자를 찾고 있습니다. Python과 Django 경험이 필요합니다."
vector = embed_text(text)

print(f"임베딩 벡터 차원: {len(vector)}")
print(f"벡터 샘플 (처음 10개): {vector[:10]}")
```

```python
# 2. Pinecone에서 유사 문서 검색
query = "프론트엔드 개발자"
results = retrieve_from_pinecone(query, k=5, index="korea-jd-dev")

print(f"\n검색 결과: {len(results)}개 문서\n")
for i, doc in enumerate(results, 1):
    print(f"[{i}] Score: {doc['score']:.4f}")
    print(f"    Content: {doc['content'][:150]}...")
    print(f"    Metadata: {doc['metadata']}")
    print()
```
