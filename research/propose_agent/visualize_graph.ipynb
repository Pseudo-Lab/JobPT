```python
from langchain_core.runnables.graph import Graph, Node, Edge, MermaidDrawMethod
from IPython.display import Image, display

#   
nodes = {
    "JD ": Node(id="JD ", name="JD ", data={"description": "Job Description   "}, metadata={"info": "metadata"}),
    "  ": Node(id="  ", name="  ", data={"description": "   "}, metadata={"info": "metadata"}),
    " ": Node(id=" ", name=" ", data={"description": " "}, metadata={"info": "metadata"}),
    " ": Node(id=" ", name=" ", data={"description": "  "}, metadata={"info": "metadata"}),
    "ATS ": Node(id="ATS ", name="ATS ", data={"description": "ATS  "}, metadata={"info": "metadata"}),
    "JD ": Node(id="JD ", name="JD ", data={"description": "JD   "}, metadata={"info": "metadata"}),
}

edges = [
    Edge(source="JD ", target="  ", data={"description": "JD     "}),
    Edge(source="  ", target=" ", data={"description": "    "}),
    Edge(source=" ", target=" ", data={"description": "   "}),
    Edge(source=" ", target="ATS ", data={"description": "   ATS "}),
    Edge(source="ATS ", target="JD ", data={"description": "ATS   "}),
]

# Graph  
app = Graph(nodes=nodes, edges=edges)

#  
display(Image(app.draw_mermaid_png(draw_method=MermaidDrawMethod.API)))
```
